```python
import pandas as pd
import numpy as np
import os

from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import colors 
from scipy.stats import pearsonr

import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
```

```python
data = pd.read_table('/share/pub/zhangyr/database/UKB/ID-field-20241226.txt',sep='\t',header=0)
eye = pd.read_csv('/share/pub/zhangyr/database/UKB/eye-ID-field.csv',header=0)
```

```python
res = {}
for col in data.columns:
    #ids = data[col]
    inter_id = set(data[col].dropna().astype(int)) & set(list(eye['Field ID'].astype(int)))
    if len(inter_id) >0:
        res['Xulab_'+col] = list(inter_id)
res
```

```python
folder = '/share/pub/zhangyr/database/UKB/output/'

out_data = {}

for f in res.keys():
    print(f)
    data = pd.read_csv(os.path.join(folder, f),index_col=0)
    pattern = 'participant\\.p'+str(res[f][0])+'_.+'
    cols_bollen = data.columns.str.contains(pattern, regex=True)
    pattern = 'participant\\.p'+str(res[f][0])
    cols_bollen = cols_bollen | data.columns.str.contains(pattern, regex=True)
    #print(fields[f][0])
    if len(res[f]) > 1:
        for i in range(1,len(res[f])):
            pattern = 'participant\\.p'+str(res[f][i])+'_.+'
            cols_bollen = cols_bollen | data.columns.str.contains(pattern, regex=True)
            pattern = 'participant\\.p'+str(res[f][0])
            cols_bollen = cols_bollen | data.columns.str.contains(pattern, regex=True)
            #print(fields[f][i])
    print('number of column: '+str(sum(cols_bollen)))
    #print(data[cols_bollen].shape)
    out_data[f] = data.loc[:,cols_bollen]
    
for x in out_data:
    out_data[x] = out_data[x].loc[out_data[x].index.notnull()]
merged_data = pd.concat(out_data.values(),axis=1)
```
