```python
def replaceCol(cols):
    new_col = []
    for c in cols:
        c = 'participant.p'+c
        c = c.replace('-','_i')
        c = c.replace('.','_a')
        new_col = new_col + [c]
    return new_col
```

```python
out_data = {}

for f in fields.keys():
    print(f)
    data = pd.read_csv(os.path.join(folder, f),index_col=0)
    pattern = 'participant\\.p'+str(fields[f][0])+'_.+'
    cols_bollen = data.columns.str.contains(pattern, regex=True)
    #print(fields[f][0])
    if len(fields[f]) > 1:
        for i in range(1,len(fields[f])):
            pattern = 'participant\\.p'+str(fields[f][i])+'_.+'
            cols_bollen = cols_bollen | data.columns.str.contains(pattern, regex=True)
            #print(fields[f][i])
    print('number of column: '+str(sum(cols_bollen)))
    #print(data[cols_bollen].shape)
    out_data[f] = data.loc[:,cols_bollen]
```
